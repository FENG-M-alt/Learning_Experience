```python
# while循环的示例+函数的定义+内部变量转全局变量
num_line = 1 # 全局变量
def multiplication_table_while():
    global num_line
    while num_line <= 9:
        num_column = 1
        while num_column <= num_line:
            print(f"{num_column}*{num_line}={num_line * num_column} ",end='')
            num_column += 1
        num_line += 1
        print()
multiplication_table_while()
```

```python
# 继承示例
class Animal:
    def speak_1(self):
        print("Animal speaks")
class Dog(Animal): # 单继承, 多继承写法: class Dog(Animal, AnotherClass):
    def speak_2(self): # 方法同名时，优先级时从左向右且为子类优先
        print("Dog barks")
        return 1
dog = Dog()
dog.speak_1()
dog.speak_2()
# book = Dog.speak_1()

# 复写父类方法示例
class Cat(Animal):
    def speak_1(self): # 复写父类方法
        print("Cat meows")
cat = Cat()
cat.speak_1()
```

```python
# 魔术方法示例
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"

    def __add__(self, other):
        if isinstance(other, Person):
            return Person(self.name + " & " + other.name, self.age + other.age)
        return NotImplemented
    
    def __it__(self, other):
        if isinstance(other, Person):
            self.age += other.age
            return self
        return NotImplemented

    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.age == other.age
        return NotImplemented

p1 = Person("Alice", 30)
p2 = Person("Bob", 25)
print(p1)  # 调用 __str__ 方法
print(p1.name)
p3 = p1 + p2  # 调用 __add__ 方法
print(p3)  # 输出合并后的 Person 对象
p1 += p2  # 调用 __it__ 方法
print(p1)  # 输出更新后的 p1 对象
print(p1 == p2)  # 调用 __eq__ 方法，比较 p1 和 p2 是否相等
```

```python
# 基础类型注解示例
var_1: int = 10
var_2: str = "Hello"
var_3: float = 3.14
var_4: bool = True
var_5: list = [1, 2, 3]
var_6: dict = {"name": "Alice", "age": 25}
var_7: tuple = (1, 2, 3)
# 类对象类型注解示例
class Person:
    pass
var_8: Person = Person()
# 容器类型详细注解示例
var_9: list[int] = [1, 2, 3]
var_10: dict[str, int] = {"one": 1, "two": 2}
var_11: tuple[str, int, float] = ("Alice", 30, 5.5)
# 函数类型注解示例
def greet(name: str) -> str:
    return f"Hello, {name}!"
print(greet("菲林斯"))

from typing import Union
def process_value(value: Union[int, str]) -> str:
    if isinstance(value, int):
        return f"Processing integer: {value}"
    elif isinstance(value, str):
        return f"Processing string: {value}"
    return "Unsupported type"
print(process_value(10))

my_list: list[Union[int, str]] = [1, "two", 3, "four"]
my_dict: dict[str, Union[int, str]] = {"one": 1, "two": "two"}
print(my_list)
print(my_dict)
```

```python
# 多态和抽象类示例
class Animal:
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        return "Woof!"
class Cat(Animal):
    def speak(self):
        return "Meow!"
animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())

def speak_animal(animal: Animal):
    print(animal.speak())
speak_animal(Dog())
speak_animal(Cat())
```
